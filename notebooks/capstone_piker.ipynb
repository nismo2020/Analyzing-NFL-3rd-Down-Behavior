Concatenating all 8 dataframes

```python
pbp_all_years = pd.concat([pbp_2013, pbp_2014, pbp_2015, pbp_2016, pbp_2017, pbp_2018, pbp_2019, pbp_2020])
pbp_all_years.head()
```

```python
pbp_all_years = pbp_all_years.loc[:, ~pbp_all_years.columns.str.contains('^Unnamed')]
```

```python
pbp_all_years['SeasonYear'] = pbp_all_years['SeasonYear'].fillna(0).astype(np.int64)
```

```python
pbp_all_years['Yards'] = pbp_all_years['Yards'].fillna(0).astype(np.int64)
```

```python
pbp_all_years.info()
```

```python
def Distance(c):
  if c['ToGo'] <= 3:
    return 'Short'
  elif c['ToGo'] <= 6:
    return 'Medium'
  elif 0.0 < c['ToGo'] < 11:
    return 'Long'
  else:
    return 'Distant'
```

```python
pbp_all_years['Distance'] = pbp_all_years.apply(Distance, axis=1)
```
